```python
import pandas as pd

df = pd.read_csv("winemag-data-130k-v2.csv")
```

```python
df.drop(["Unnamed: 0"], axis = 1, inplace = True)
```

```python
df = df.dropna()

df["points"] = df["points"].astype('object', copy=False)
df["price"] = df["price"].astype('object', copy=False)
```

```python
from time import time
```

```python
def export_wines():
    for i in range(1000):
        id = int(time())+i
        db.child("wines").child(id).set(df.iloc[i].to_dict())
```

```python
export_wines()
```

```python
def return_top15():
    top_wines = db.child("wines").order_by_child("points").limit_to_last(15).get()
    return top_wines.val()
```

```python
return_top15()
```

```python
def vinhos_province(provincia):
    wines_province = db.child("wines").order_by_child("province").equal_to(provincia).get()
    return wines_province.val()
```

```python
vinhos_province("California")
```

```python
def remover_vinhos(timestamp):
        db.child("wines").child(timestamp).remove()
        return "Vinho removido com sucesso!"
```

```python
def atualizar_vinhos(timestamp, chave, valor):
    db.child("wines").child(timestamp).update({chave: valor})
    return "O usuario ", timestamp, " foi atualizado com sucesso!"
```
